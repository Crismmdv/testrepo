```python
import pandas as pd
```

```python
dicc= pd.read_excel('../Anexo_9.xlsx', sheet_name='DICT')
tab1= pd.read_excel('../Anexo_9.xlsx', sheet_name='Anexo 9',skiprows=[])
#display(tab1)
```

```python
tab3=tab1.copy()
dicc2=dicc.copy()
col=list(dicc2.B.dropna())
new=list(dicc2.A.dropna())
old=col
nomb=dict(zip(old,new))
tab2=tab1[col]

tab2.rename(columns=nomb,inplace=True)
#display(nomb)
tab3=tab3.astype({'Fecha':str})
tab2=tab2.astype({'Fecha muestreo':str})
#display(tab3.Fecha)
for i in tab3.index:
    tab3.at[i,'Fecha']=tab3.at[i,'Fecha'][8:11]+'-'+tab3.at[i,'Fecha'][5:7]+'-'+tab3.at[i,'Fecha'][0:4]
for y in tab3.index:
    tab2.at[y,'Fecha muestreo']=tab2.at[y,'Fecha muestreo'][8:11]+'-'+tab2.at[y,'Fecha muestreo'][5:7]+'-'+tab2.at[y,'Fecha muestreo'][0:4]
    tab2.at[y,'orden']=tab2.at[y,new[0]][-4:]
tab2=tab2.astype({'orden':int})
#tab2.sort_values(by='orden',inplace=True)
tab2.reset_index(inplace=True,drop=True)
tab2.drop(columns='orden',inplace=True)
#display(tab2.head(5))

dicc2.set_index('A',drop=True,inplace=True)
tab2.set_index(new[0],drop=True,inplace=True)
tab3.set_index(old[0],drop=True,inplace=True)
tab2.rename_axis(axis=1,mapper=new[0],inplace=True)
tab3.rename_axis(axis=1,mapper=old[0],inplace=True)
#display(tab2.head(5))
tab2.fillna('---',inplace=True)
with pd.ExcelWriter('../Anexo_9_out.xlsx') as writer:
    tab3.to_excel(writer, sheet_name='Anexo 9')
    dicc2.to_excel(writer, sheet_name='DICT')
    tab2.T.to_excel(writer, sheet_name='nuevo')
```
